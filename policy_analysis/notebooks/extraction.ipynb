```python
db_client = database_client.DatabaseClient()
```

```python
tables = db_client.list_tables()
print("Tables disponibles :", tables)
```

```python
table_info = db_client.get_table_info("policies_abstracts_all")
table_info
```

```python
query = """
SELECT
    am.title,
    am.abstract,
    am.studied_sector AS studied_sector,
    am.year_of_publication,
    am.doi,
    paa.extracted_data
FROM article_metadata am
LEFT JOIN policies_abstracts_all paa ON am.doi = paa.doi
WHERE am.sufficiency_mentioned = False
ORDER BY RANDOM()
LIMIT 50;
"""
results = db_client.execute_query(query)

df = pd.DataFrame(results)

df.to_csv('random_articles.csv', index=False)
df
```
